```python
import matplotlib.pyplot as plt
import pickle

# Load the training histories from the pickle files
with open('MaskClassification-DenseNet.pkl', 'rb') as file:
    history_densenet = pickle.load(file)

with open('MaskClassification-MobileNet.pkl', 'rb') as file:
    history_mobilenet = pickle.load(file)

with open('MaskClassification-Resnet.pkl', 'rb') as file:
    history_resnet = pickle.load(file)

with open('MaskClassification-VGG19.pkl', 'rb') as file:
    history_vgg19 = pickle.load(file)
```

```python
# Determine the number of epochs
epochs = range(1, len(history_densenet.history['accuracy']) + 1)

# Plot training & validation accuracy values
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(epochs, history_densenet.history['accuracy'], label='DenseNet')
plt.plot(epochs, history_mobilenet.history['accuracy'], label='MobileNet')
plt.plot(epochs, history_resnet.history['accuracy'], label='ResNet')
plt.plot(epochs, history_vgg19.history['accuracy'], label='VGG19')
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.xticks(epochs)
plt.legend(loc='upper left')

plt.subplot(1, 2, 2)
plt.plot(epochs, history_densenet.history['val_accuracy'], label='DenseNet')
plt.plot(epochs, history_mobilenet.history['val_accuracy'], label='MobileNet')
plt.plot(epochs, history_resnet.history['val_accuracy'], label='ResNet')
plt.plot(epochs, history_vgg19.history['val_accuracy'], label='VGG19')
plt.title('Model validation accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.xticks(epochs)
plt.legend(loc='upper left')

plt.show()

# Plot training & validation loss values
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(epochs, history_densenet.history['loss'], label='DenseNet')
plt.plot(epochs, history_mobilenet.history['loss'], label='MobileNet')
plt.plot(epochs, history_resnet.history['loss'], label='ResNet')
plt.plot(epochs, history_vgg19.history['loss'], label='VGG19')
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.xticks(epochs)
plt.legend(loc='upper left')

plt.subplot(1, 2, 2)
plt.plot(epochs, history_densenet.history['val_loss'], label='DenseNet')
plt.plot(epochs, history_mobilenet.history['val_loss'], label='MobileNet')
plt.plot(epochs, history_resnet.history['val_loss'], label='ResNet')
plt.plot(epochs, history_vgg19.history['val_loss'], label='VGG19')
plt.title('Model validation loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.xticks(epochs)
plt.legend(loc='upper left')

plt.show()
```
